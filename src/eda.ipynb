```python
# creating a new column Profit Margin in the dataframe
df["Profit Margin"] = df["Profit"] / df["Sales"]
```

```python
# created another column Sales per item in the dataframe
df["Sales per item"] = df["Sales"] / df["Quantity"]
```

```python
# displays the first 5 rows 
df.head()
```

```python
# Convert the 'Category' column into one-hot encoded dummy variables with integer values (0 and 1).
df = pd.get_dummies(df,columns=["Category"] , dtype=int)
```

```python
# displays the first 5 rows 
df.head()
```

```python
# importing StandardScaler class 
from sklearn.preprocessing import StandardScaler 
num_cols = ["Sales","Profit","Quantity","Discount"]
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# display first five rows 
df.head()
```

```python
#importing train_test_split function
from sklearn.model_selection import train_test_split
#we will predict profit 
x = df.drop("Profit", axis=1)
y = df["Profit"]
```

```python
## Divide data into training and testing sets.
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
import matplotlib.pyplot as plt

# plotting bar chart of sales by sub_category
sales_by_subcategory = df.groupby("Sub-Category")["Sales"].sum()

sales_by_subcategory.plot(kind="bar", figsize=(10,5))

plt.title("Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

## 📊 Sales by Sub-Category

This bar chart shows the total sales for each product sub-category in the dataset. It helps compare sales performance across different categories and identify the highest- and lowest-performing sub-categories.

### Key Insights
- **Chairs** recorded the highest total sales.
- **Machines**, **Copiers**, **Phones**, and **Tables** also contributed significantly to total sales.
- Some sub-categories show negative values, indicating possible returns, losses, or data inconsistencies that should be investigated.
- This visualization helps identify which product sub-categories perform well and which may require further analysis.

```python
# Average Sales by Sub-Category
avg_sales = df.groupby("Sub-Category")["Sales"].mean().reset_index()

# Optional: Sort by average sales
avg_sales = avg_sales.sort_values(by="Sales")

# Create Line Chart
plt.figure(figsize=(12,6))
plt.plot(avg_sales["Sub-Category"], avg_sales["Sales"],
         marker='o', linestyle='-', linewidth=2)

plt.title("Average Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Average Sales")
plt.xticks(rotation=90)
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 📈 Average Sales by Sub-Category

This line chart displays the average sales for each product sub-category. It helps identify which sub-categories have the highest and lowest average sales and provides a clear comparison of their overall performance.

### Key Insights
- **Copiers** have the highest average sales among all sub-categories.
- **Machines** and **Tables** also show relatively high average sales.
- **Bookcases**, **Chairs**, and **Phones** have moderate average sales.
- **Fasteners**, **Art**, **Labels**, and **Paper** have the lowest average sales.
- This visualization helps businesses understand which product sub-categories generate higher average revenue and supports better inventory and sales planning.
