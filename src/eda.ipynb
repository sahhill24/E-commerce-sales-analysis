### Task 3: Distribution Analysis

#### Observation 1: Sales Distribution
- Most orders have low sales values.
- Only a few orders have very high sales, indicating some high-value purchases.
- The sales distribution is right-skewed.

#### Observation 2: Profit Distribution
- Most orders generate small profits close to zero.
- A few orders have negative profit, which indicates losses.
- Very high profits are rare.

#### Observation 3: Quantity Distribution
- Most customers purchase between 1 and 4 items per order.
- Orders with large quantities (10 or more items) are uncommon.
- The quantity distribution is also right-skewed.

### Task 4: Key Insights

#### Insight 1
Most orders have low sales values, while only a small number of orders contribute to very high sales. This indicates that high-value purchases are relatively rare.

#### Insight 2
Most orders generate positive profits, but some orders result in losses. This suggests that certain products or discounts may reduce profitability.

#### Insight 3
Customers usually purchase small quantities (1–4 items) per order. Large-quantity purchases are uncommon, indicating that most transactions involve only a few products
